Q7 Pivot Table Creation

Create a pivot table showing:

  > Rows → Region

  > Columns → Product

  > Values → Total Sales

Which region sells the most laptops?



```
# Create Pivot Table
pivot_table = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Product",
    aggfunc="sum"
)

# Display Pivot Table
print(pivot_table)

# Find which region sells the most laptops
top_laptop_region = pivot_table["Laptop"].idxmax()
top_laptop_sales = pivot_table["Laptop"].max()

print("\nRegion with Highest Laptop Sales:", top_laptop_region)
print("Total Laptop Sales:", top_laptop_sales)
```

Q8 Profitability Analysis

Calculate the average profit for each product category.

Which category is most profitable?



```
# Calculate average profit for each category
avg_profit = df.groupby("Category")["Profit"].mean()

# Display average profit
print(avg_profit)

# Find the most profitable category
most_profitable = avg_profit.idxmax()
highest_profit = avg_profit.max()

print("\nMost Profitable Category:", most_profitable)
print("Average Profit:", highest_profit)
```

Q9 Discount Impact Study

Analyze the relationship between discount and profit.

Tasks:

  > Calculate the average profit for each discount level

  > Identify whether higher discounts lead to lower profit



```
# Calculate average profit for each discount level
discount_profit = df.groupby("Discount")["Profit"].mean()

# Display result
print("Average Profit for Each Discount Level:\n")
print(discount_profit)

# Optional: Sort by discount
print("\nSorted by Discount:\n")
print(discount_profit.sort_index())

# Check relationship between discount and profit
correlation = df["Discount"].corr(df["Profit"])

print("\nCorrelation between Discount and Profit:", correlation)
```

Q10 End-to-End Data Wrangling Task

Perform the following workflow:

  > Filter the dataset where Sales > 3000

  > Group data by Region and Product

  > Calculate total sales

  > Reshape the result into a pivot table

Explain the business insights obtained from this analysis.



```
# Step 1: Filter dataset where Sales > 3000
filtered_data = df[df["Sales"] > 3000]

# Step 2: Group data by Region and Product
grouped_data = filtered_data.groupby(
    ["Region", "Product"]
)["Sales"].sum()

# Display grouped result
print("Grouped Sales Data:\n")
print(grouped_data)

# Step 3: Reshape into Pivot Table
pivot_table = grouped_data.unstack()

# Display Pivot Table
print("\nPivot Table:\n")
print(pivot_table)
```
