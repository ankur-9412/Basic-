A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?


```
# Mean is the most representative value of their weekly sales

   The mean is the most representative value here because:

*   The data is fairly balanced with no extreme outliers.
*   Sales gradually increase and vary normally through the week.
*   The mean uses all observations, giving a good overall picture of weekly sales.

          Mean= 10+12+11+15+14+13+12/7 = 87/7​≈12.43
```

# Q3: Mode in Real Life (Easy)
A store records the shoe sizes sold in one day: [7, 8, 9, 8, 8, 10, 7, 9].
What is the mode, and why is this information useful for the store manager?


```
The mode is the value that occurs most frequently in a data set.

In the given data:

7 → 2 times  8 → 3 times  9 → 2 times  10 → 1 time

So, the mode is 8.

Why this is useful for the store manager:
  *   Shoe size 8 is the most commonly sold size.
  *   The manager can stock more of size 8 to meet customer demand.
  *   It helps reduce shortages and increase sales by focusing on the most popular size.
  *   the mode helps in better inventory planning and decision-making.
```

```
Median = $10,200
Why the median is better than the mean:
*   The price $50,000 is an extreme value (outlier) compared to the other car prices.
*   The mean would be pulled upward by this very expensive car and would not    reflect the typical used car price.
*   
The median is not affected by outliers, so it gives a more realistic idea of what most cars cost.
```

#Q7: Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).

  *Company A: variance = 6

  *Company B: variance = 15

 Which company is more consistent, and why?


```
Company A is more consistent.
Reason:
* Variance measures how much the delivery times spread out from the average.
* A lower variance means the delivery delays are more tightly clustered around the mean.
* Company A has a variance of 6, while Company B has a variance of 15.

Since 6 < 15, Company A’s delivery times show less variability, making it more reliable and consistent than Company B.
```

Q8: Standard Deviation in Context (Hard)
A finance student compares the daily price fluctuations of two cryptocurrencies.

Coin A: standard deviation = $30

Coin B: standard deviation = $120

Which coin is riskier to invest in, and why?


```
Coin B is riskier to invest in.

Why:

* Standard deviation measures how much prices typically fluctuate around the average price.
* A higher standard deviation means greater volatility and higher risk.
* Coin A has a standard deviation of $30, showing relatively smaller daily price changes.
* Coin B has a standard deviation of $120, meaning its price changes are much larger and more unpredictable.

Since $120 > $30, Coin B is riskier, as its price is more volatile and less stable for investment.
```
#Q9: Combining Measures (Hard)
A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].
Find the mean and standard deviation. What do these values together tell you about the family’s energy use
pattern?



```
     Mean = 400+420+390+450+410/5 = 2070/5 = 414 kWh
     
Value	     Deviation  	Squared Deviation
400-414       −14         	196
420-414        +6           36
390-414       −24	          576
450-414	      +36           1296
410-414       −4            16

Sum of squared deviations:
 196+36+576+1296+16=2120

Variance:

  2120/5 = 424

Standard deviation:
       sqrt(424) ≈ 20.6kWh
Standard deviation ≈ 20.6 kWh

What do these values tell us together?
* The mean (414 kWh) shows the family’s typical monthly electricity usage.
*The standard deviation (≈ 20.6 kWh) is relatively small compared to the mean.
*This indicates that electricity usage is fairly consistent each month, with only moderate fluctuations.
There are no extreme spikes or drops in consumption.

```
