Q1. Create a countplot to show the number of passengers by gender.


```

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Downloads/Titanic.csv')

# Display first 5 rows
print(df.head())

# Create countplot
plt.figure(figsize=(6,5))
sns.countplot(x='Sex', data=df)

# Add title and labels
plt.title("Number of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

# Show plot
plt.show()
```

Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.



```
# Count passengers in each class
class_counts = df['Pclass'].value_counts().sort_index()

# Create bar chart
plt.figure(figsize=(7,5))
plt.bar(class_counts.index, class_counts.values,
        color=['skyblue', 'orange', 'green'])

# Add title and labels
plt.title("Number of Passengers in Each Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

# Set x-axis labels
plt.xticks([1, 2, 3], ['Class 1', 'Class 2', 'Class 3'])

# Show plot
plt.show()
```

Q3. Create a histogram to show the distribution of passenger age.



```
# Handle missing values in Age column
df['Age'].fillna(df['Age'].median(), inplace=True)

# Create histogram
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')

# Add title and labels
plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

# Show plot
plt.show()
```

5. Create a scatterplot to

 a. show the relationship between age and fare.

 b. analyze the relationship between family size (SibSp + Parch) and fare.



```
# Q5(a) Scatterplot: Relationship between Age and Fare.

# Handle missing values in Age column
df['Age'].fillna(df['Age'].median(), inplace=True)

# Create scatterplot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', data=df)

# Add title and labels
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

# Show plot
plt.show()


# Q5(b) Scatterplot: Relationship between Family Size and Fare

# Create Family Size column
df['FamilySize'] = df['SibSp'] + df['Parch']

# Create scatterplot
plt.figure(figsize=(8,5))
sns.scatterplot(x='FamilySize', y='Fare', data=df)

# Add title and labels
plt.title("Relationship Between Family Size and Fare")
plt.xlabel("Family Size (SibSp + Parch)")
plt.ylabel("Fare")

# Show plot
plt.show()
```

Q6. Create a countplot to show survival count by gender.



```
# Create countplot
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df)

# Add title and labels
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

# Add legend labels
plt.legend(title="Survived", labels=["No", "Yes"])

# Show plot
plt.show()

```

Q8. Create a heatmap to show correlation between numerical variables


```
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64'])

# Create correlation matrix
correlation_matrix = numerical_cols.corr()

# Create heatmap
plt.figure(figsize=(10,7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title("Correlation Heatmap of Numerical Variables")

# Show plot
plt.show()
```

Q9. Identify which passenger class paid the highest fare on average.


```
# Calculate average fare by passenger class
avg_fare = df.groupby('Pclass')['Fare'].mean()

# Display average fare
print(avg_fare)

# Create bar chart
plt.figure(figsize=(7,5))
sns.barplot(x=avg_fare.index, y=avg_fare.values)

# Add title and labels
plt.title("Average Fare Paid by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

# Set class labels
plt.xticks([0, 1, 2], ['Class 1', 'Class 2', 'Class 3'])

# Show plot
plt.show()
```
