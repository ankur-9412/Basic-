#Question 1: A die is rolled. What is the probability of getting:

(a) An even number

(b) A number greater than 4


```
(a) Probability of getting an even number
     Even numbers on a die = {2, 4, 6}
     Number of favorable outcomes = 3
              P(even number)=3/6 = 1/2

(b) Probability of getting a number greater than 4

Numbers greater than 4 = {5, 6}
Number of favorable outcomes = 2
         P(number > 4)=2/6 = 1/3

```
# Question 2: In a class of 50 students:

 * 20 like Mathematics (M)

* 15 like Science (S)

* 5 like both subjects

What is the probability that a student chosen at random likes Mathematics or Science?



```
 Given:

 Total number of students = 50
 Students who like Mathematics (M) = 20
 Students who like Science (S) = 15
 Students who like both subjects (M ∩ S) = 5
     
     Using the Inclusion–Exclusion Principle:
               n(M∪S)=n(M)+n(S)−n(M∩S)
                 n(M∪S)=20+15−5=30
    Probability Calculation:
                      P(M or S)=30/50 = 3/5
```
# Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability
that the next ball is also red (without replacement)?


```
Given:
A bag has 3 red and 2 blue balls (total = 5).

One ball is drawn and it is red.
Now we need the probability that the next ball is also red, without replacement.

    Step-by-step Solution

After drawing 1 red ball, remaining balls:

Red balls = 2
Blue balls = 2
Total balls = 4
             P(next ball is red)=Remaining red balls/Total remaining balls = 2/4 =1/2
	​

```
# Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?


```
Stratified Sampling
   Reason (Simple Explanation)

The school population has:
60% boys
40% girls
If we want equal representation of both genders in the sample, then Stratified Sampling is the best method.

Why Stratified Sampling?

 * In Stratified Sampling, the population is divided into groups (called strata) — here, boys and girls.
  * then, samples are taken separately from each group in the required proportion.
  * This ensures both boys and girls are properly represented, even if their population percentages are different.

Why NOT Simple Random Sampling?

 * In Simple Random Sampling, selection is completely random.
 * It may result in more boys and fewer girls, or vice versa.
 * So, equal representation is not guaranteed.

```
# Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height= 158 cm.

Find the sampling error.


```
Given:

Population mean height (μ) = 160 cm

Sample mean height (x̄) = 158 cm

     Sampling Error Formula:
                 Sampling Error=xˉ−μ
                    =158−160=−2 cm
                    
```
#Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees,
what
is the standard error of the mean (SEM)?



```
Given:

    Population standard deviation (σ) = ₹5,000
    Sample size (n) = 100
          
           Formula for Standard Error of Mean (SEM):
                        SEM=​σ/sqrt(n)
           Calculation:
                   SEM=5000/sqrt(100) = 5000/10 = 500
                   
```
#Question 7: In a group of 100 students:

40 like Cricket (C)

30 like Football (F)

10 like both Cricket and Football

Find the probability that a student likes at least one sport.



```
Given:

Total students = 100
Like Cricket (C) = 40
Like Football (F) = 30
Like both (C ∩ F) = 10

Students who like at least one sport (C ∪ F):
                n(C∪F)=n(C)+n(F)−n(C∩F)
                       =40+30−10=60

            Probability:
                  P(at least one sport)=60/100 = 0.6

```

#Question 10: Differentiate between discrete and continuous random variables with examples.


```
| Basis   | Discrete        | Continuous         |
| ------- | --------------- | ------------------ |
| Values  | Countable       | Uncountable        |
| Nature  | Separate values | Any value in range |
| Example | Number of cars  | Height, weight     |

A continuous random variable can take any value within a given range.

Examples:

   * Height of students (160.2 cm, 160.25 cm, etc.)

   * Weight of a person

   * Time taken to complete a task
```
