3.Write the end-to-end analytics workflow in the correct order and briefly explain each step.
     ->
      1. Define the Problem
         Identify what question needs to be answered or what issue needs to be solved.
      
      2. Data Collection and Understanding
      Gather data from sources like databases, Excel files, or APIs, and understand what each data field represents.

      3. Data Preparation
      Clean and organize the data—remove duplicates, handle missing values, and format data properly.
      Clean data = reliable results.
      4. Data Analysis
      Apply statistical techniques, queries, or models to find patterns, trends, and insights in the data.
      5.Data Visualization
      Present findings using charts, graphs, or dashboards (Power BI / Tableau) so others can easily understand the results.

4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?
      -> Prompt Engineering is about asking the AI the right way to get the best answer.
        1. Instruction Prompt
          This type of prompt clearly tells the AI what to do.
          It usually starts with action words like explain, write, summarize, analyze, etc.
       2. Role-Based Prompt
           In this prompt, you assign a role to the AI so it responds in that style or tone.

5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles,responsibilities, and focus?
    
    1. Focus Area
       Business Analyst: Focuses on understanding business needs and improving processes.
       Data Analyst: Focuses on analyzing data to find trends and insights.

    2. Main Goal
       Business Analyst: Helps the organization decide what actions or changes should be made.
       Data Analyst: Helps the organization make decisions based on data.
    3. Key Responsibilities
       Business Analyst: Collects business requirements, communicates with stakeholders, and suggests solutions.
       Data Analyst: Collects, cleans, analyzes data and creates reports or dashboards.

7.Give three applications of Data Analytics across different industries and explain how it creates value.


        1. Healthcare
         Application: Analyzing patient records, test results, and treatment outcomes.
         Value Created: Helps doctors identify diseases earlier, choose better treatment plans, and reduce medical errors.
         Example: Predicting which patients are at risk of diabetes or heart disease.
         2. Retail / E-commerce
         Application: Studying customer buying patterns, product demand, and sales trends.
         Value Created: Helps companies offer personalized recommendations, manage inventory better, and increase sales
         Example: Showing “Recommended Products” based on previous purchases.
         3. Banking & Finance
         Application: Monitoring transactions to detect fraud, credit scoring, and risk analysis.
         Value Created: Improves security, reduces financial loss, and helps banks decide who is eligible for loans.
         Example: Detecting unusual spending to prevent fraud alerts.  

8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.
     
      Descriptive Analytics
        Purpose: Understand what has happened in the past.
        Tools/Methods: Reports, dashboards, summaries.

      Diagnostic Analytics
         Purpose: Understand why something happened.
         Tools/Methods: Data drilling, correlation analysis, root cause analysis.

      Predictive Analytics
          Purpose: Predict what is likely to happen in the future.
          Tools/Methods: Machine learning models, forecasting.

      Prescriptive Analytics
          Purpose: Suggest what actions should be taken.
          Tools/Methods: Optimization models, decision algorithms.

      Generative AI Analytics (Latest Stage)
          Purpose: Not just analyze data, but create new content, insights, and solutions.
          Tools/Methods: LLMs like GPT, AI copilots, AutoML.
