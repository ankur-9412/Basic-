Question 1 : Create a Card Visual to show Total Sales.


```
1. Insert a Card Visual
2. Drag the Sales field into the Values section
3.Ensure aggregation is set to Sum
4.Format the title as “Total Sales”

```

Question 2  : Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the
highest sales?


```
1.Insert a Donut Chart
2.Drag:
   > Ship Mode → Legend
   > Sales → Values (Sum)

3. Turn on Data Labels
4. The largest slice represents the ship mode generating the highest sales.
 In most Superstore datasets, Standard Class generates the highest sales.
```
