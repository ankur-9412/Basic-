Q8. What is the difference between a Worksheet and a Dashboard in Tableau?

Why is a dashboard more useful for decision-makers?

```
Difference:
🔹 Worksheet:
     > A Worksheet is a single view or chart in Tableau
     > It is used to create one visualization at a time (e.g., bar chart, line chart)
     > Focuses on analyzing one aspect of data
🔹 Dashboard:
     > A Dashboard is a collection of multiple worksheets combined in one screen
     > It shows multiple visualizations together
     > Used to provide a complete view of data
  
Why Dashboard is More Useful for Decision-Makers:
    > It provides all important insights in one place
    > Helps in quick and better decision-making
    > Allows comparison of multiple charts at once
    > Interactive features (filters, actions) make analysis easier
```
Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)


```
Tableau Prep is a data preparation tool used to clean, combine, and transform raw data before analysis in Tableau.

Why is it Important in Real-World Data Projects?
  > Real-world data is often messy and inconsistent
  > Tableau Prep helps in cleaning and organizing data
  > Improves data accuracy and reliability
  > Saves time by automating data preparation tasks

Data Cleaning Steps in Tableau Prep
 🔹 1. Remove Null Values
Steps:
  > Open the dataset in Tableau Prep
  > Go to the field where null values exist
  > Click on the null values (shown as “Null”)
  > Choose “Filter Data” or uncheck null values

2. Remove Duplicate Records
Steps:
  > Add a Clean Step
  > Select all relevant fields
  > Use the option “Remove Duplicates” (or use Group & Keep one record)

3. Change Incorrect Data Types
Steps:
  > Identify the field with incorrect data type
  > Click on the data type icon (e.g., ABC, #, calendar icon)
  > Change it to the correct type:
           > Text → String
           > Number → Integer/Decimal
           > Date → Date
```
