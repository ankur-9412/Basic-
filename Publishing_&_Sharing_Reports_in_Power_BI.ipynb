# Easy Level
Q1. Explain why publishing and sharing reports is considered more important than just building dashboards
in Power BI Desktop.
```
1. Collaboration:
A dashboard created in Power BI Desktop can only be viewed by the person who created it. When it is published to Power BI Service, it can be shared with team members and stakeholders.

2. Better Decision Making:
Managers and business users need access to reports to understand data trends and make informed decisions. Sharing reports allows them to view and use these insights.

3. Accessibility:
After publishing, reports can be accessed from web browsers or mobile devices, allowing users to view them anytime and from anywhere.

4. Automatic Data Refresh:
In Power BI Service, data can be scheduled for automatic refresh, ensuring that reports always display the latest information.

5. Security and Permissions:
Reports can be shared with specific users or groups, and permissions can be controlled to ensure data security.
```
Q2. Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.


```
Power BI Service is a cloud-based platform provided by Microsoft that allows users to publish, share, and manage reports and dashboards online. It is accessed through a web browser and helps organizations collaborate and distribute data insights.

Role in Comparison to Power BI Desktop

1. Power BI Desktop (Report Creation Stage):
  > It is a desktop application used to connect to data sources, clean and
    transform data, create data models, and design reports and dashboards.

  > It is mainly used by data analysts and developers for building reports.

2. Power BI Service (Sharing and Collaboration Stage):
  > After reports are created in Desktop, they are published to Power BI Service.

  > In the Service, reports can be shared with team members, viewed online, and accessed from different devices.


```

Q3. What does publishing a report mean in Power BI? List any two outcomes that occur after a report is
published.

```
Publishing a report in Power BI means uploading a report created in Microsoft Power BI Desktop to the cloud platform Microsoft Power BI Service so that it can be accessed and shared online.

Two Outcomes After Publishing
1. Online Access:
  > The report becomes available in Power BI Service, allowing users to view it through a web browser or mobile device.

2. Sharing and Collaboration:
  > The report can be shared with other users or team members, enabling them to interact with the visuals and use the insights for decision-making.

```

# Medium Level

Q4. List and explain the steps involved in publishing a report from Power BI Desktop to Power BI Service in the
correct sequence.



```
1. Create and Save the Report
   > First, create the report in Power BI Desktop by connecting to data, transforming it, and designing visuals. Then save the report file (.pbix).

2. Sign In to Power BI
  > Open Power BI Desktop and sign in using your Power BI account to connect with the Power BI Service.

3. Click the Publish Option
  > Go to the Home tab in Power BI Desktop and click the Publish button.

4. Select a Workspace
   > A window will appear asking where to publish the report. Choose a workspace (such as My Workspace or another shared workspace).

5. Upload the Report
  > After selecting the workspace, Power BI will upload the report and dataset to Power BI Service.

6. View the Published Report
  > Once publishing is complete, open Power BI Service in a browser to view, share, or manage the report online.
```
Q5. Differentiate between Report and Dashboard in Power BI based on structure and usage.


```
| Basis             | Report                                                                                     | Dashboard                                                                                 |
| ----------------- | ------------------------------------------------------------------------------------------ | ----------------------------------------------------------------------------------------- |
| **Structure**     | A report contains **multiple pages** with many visuals such as charts, tables, and graphs. | A dashboard is **a single-page canvas** that shows key visuals.                           |
| **Creation Tool** | Reports are mainly created in **Microsoft Power BI Desktop**.                              | Dashboards are created in **Microsoft Power BI Service**.                                 |
| **Data Source**   | Usually based on **one dataset**.                                                          | Can display visuals from **multiple reports or datasets**.                                |
| **Usage**         | Used for **detailed analysis and exploration of data**.                                    | Used for **quick overview of important metrics (KPIs)**.                                  |
| **Interactivity** | Users can use **filters, slicers, and drill-down features**.                               | Interaction is limited; mainly used for **monitoring important information at a glance**. |

```
Q6. Explain the difference between the following three sharing methods in Power BI Service:
  > Direct Report Sharing
  > Workspace Access
  > Power BI Apps
Also mention one use case for each.


```
1. Direct Report Sharing
Explanation:
Direct report sharing allows a user to share a specific report or dashboard directly with another user by entering their email address. The shared user can open and interact with the report in Power BI Service.

Use Case:
When a manager wants to share a single report with a few team members for quick review.

2. Workspace Access
Explanation:
A workspace is a collaborative environment where multiple users can create, edit, and manage reports, datasets, and dashboards together. Different roles like Admin, Member, or Viewer can be assigned.

Use Case:
When a data analytics team works together to develop and manage multiple reports for a project.

3. Power BI Apps
Explanation:
Power BI Apps allow organizations to package dashboards and reports from a workspace and distribute them to a large group of users in a structured way.

Use Case:
When a company wants to distribute a finalized set of reports (such as sales or performance dashboards) to many employees across the organization.
```
Q7. Describe the four workspace roles available in Power BI and explain how access control is managed using
these roles.

```
1. Admin
Explanation:
Admin has the highest level of access in the workspace. They can manage all workspace settings and permissions.

Permissions:
  > Add or remove users
  > Assign roles to members
  > Delete or modify workspace content
  > Publish and share reports

2. Member
Explanation:
Members can create, edit, and publish content in the workspace, but they cannot manage workspace settings like an Admin.

Permissions:
  > Edit reports and dashboards
  > Publish new reports
  > Share workspace content

3. Contributor
Explanation:
Contributors can create and modify content, but they cannot share the workspace or manage permissions.

Permissions:
  > Upload datasets and reports
  > Edit existing reports
  > Cannot add or remove users

4. Viewer
Explanation:
Viewers have read-only access to the workspace content.

Permissions:
  > View dashboards and reports
  > Interact with visuals and filters
  > Cannot edit or publish content

Access Control in Workspaces

Access control is managed by assigning these roles to users. Each role has different permissions, ensuring that only authorized users can create, edit, share, or manage reports, while others can only view them.
```
# Hard Level
Q8. Why are Power BI Apps considered more secure and scalable than direct sharing?
Explain this from a governance and enterprise usage perspective.


```
1. Better Governance
With Power BI Apps, reports and dashboards are published from a workspace as a packaged app. Only authorized developers or admins can update the content, which ensures consistent and controlled reporting across the organization.

2. Centralized Management
Power BI Apps allow organizations to manage reports from a single workspace. When updates are made to the app, all users automatically see the latest version, which reduces confusion caused by multiple shared copies.

3. Improved Security
Apps allow administrators to control access at the app level, ensuring that only approved users or groups can access the reports. This helps maintain data security and compliance.

4. Scalability for Large Organizations
Direct sharing is suitable for a small number of users, but Power BI Apps can distribute reports to hundreds or thousands of users efficiently, making them ideal for enterprise environments.
```
Q9. Explain Row-Level Security (RLS) in Power BI.
How does RLS ensure that different users see different data using the same report?


```
Row-Level Security (RLS) in Microsoft Power BI is a feature that restricts data access for different users based on their roles. It ensures that users can only see the data that they are authorized to view, even when they are using the same report or dashboard.

How RLS Works
1. Create Roles
  > In Microsoft Power BI Desktop, roles are created with specific filter conditions on tables. For example, a filter may allow a role to see data only for a particular region.

2. Define Data Filters
  > Each role is assigned a DAX filter expression (for example: Region = "North") that limits the rows visible to users in that role.

3. Publish the Report
  > The report is published to Microsoft Power BI Service.

4. Assign Users to Roles
  > In Power BI Service, users or groups are assigned to the roles created in Power BI Desktop.


How Different Users See Different Data
  > When a user opens the report:
     > Power BI checks which role the user belongs to.
     > The defined filters are automatically applied to the dataset.
     > As a result, the user only sees the rows allowed for that role.

Example
If a sales report contains data for North, South, East, and West regions:
    > A North Manager will only see North region data.
    > A South Manager will only see South region data.
Both users access the same report, but the data displayed is different.
```
Q10.Assume an organization shares reports without using Apps, workspace roles, or RLS.
Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain
how Power BI best practices help prevent them.



```
1. Unauthorized Data Access
Risk:
  > Without Row-Level Security (RLS), all users may see the entire dataset, including sensitive information that should be restricted.

Best Practice:
  > Implement RLS to restrict data access so that users only see the data relevant to their role or department.

2. Poor Governance and Content Management
Risk:
  > Without proper workspace roles, anyone might edit, delete, or publish reports, which can cause data inconsistencies or accidental changes.

Best Practice:
   > Use workspace roles (Admin, Member, Contributor, Viewer) to control who can create, edit, or only view reports, ensuring proper governance.

3. Data Misuse and Version Confusion
Risk:
  > If reports are shared directly with many users, multiple copies of the same report may exist, leading to confusion about which version is correct and possible misuse of outdated data.

Best Practice:
   > Use Power BI Apps to distribute official and verified reports from a central workspace so that all users access the same trusted version.
```
