4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?

Visual-Level Filter

A Visual-Level Filter is a filter that is applied to only one specific visual (chart, table, card, etc.) in a report.

| Filter Type  | Affects                 | Scope          | Example                      |
| ------------ | ----------------------- | -------------- | ---------------------------- |
| Visual-Level | One visual only         | Smallest scope | One chart shows North region |
| Page-Level   | All visuals on one page | Medium scope   | Entire page shows 2023 data  |
| Report-Level | Entire report           | Largest scope  | All pages show East region   |

6. What is the role of Drill-Down in Power BI? How does it improve data exploration?



```
Drill-Down in Power BI allows users to navigate from high-level summary data to more detailed data within a visual. It works when your data has a hierarchy (for example: Year → Quarter → Month → Day).

How Drill-Down Improves Data Exploration
 > Better Analysis – Users can identify trends and patterns more clearly.
 > Interactive Exploration – Users control how deep they want to analyze the data.
 > Detailed Insights – Helps find the root cause of changes (e.g., which month caused a drop in yearly sales).
 > Cleaner Reports – No need to create multiple separate charts for each level.
```

10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example

| Point             | Drill-Down                                                                                                   | Drill-Through                                                                                               |
| ----------------- | ------------------------------------------------------------------------------------------------------------ | ----------------------------------------------------------------------------------------------------------- |
| 1. Definition     | Drill-Down allows users to move from a higher level of data to a more detailed level within the same visual. | Drill-Through allows users to navigate from one report page to another page with more detailed information. |
| 2. Location       | It works within the same chart or visual.                                                                    | It opens a separate report page for detailed analysis.                                                      |
| 3. Data Structure | It uses hierarchical data such as Year → Quarter → Month → Day.                                              | It uses filters to show detailed data for the selected item.                                                |
| 4. User Action    | Users click on a data point to view deeper levels of the hierarchy.                                          | Users usually right-click on a data point and choose the drill-through option.                              |
| 5. Example        | Viewing sales from Year level down to Month level in the same chart.                                         | Clicking on a product name to open another page showing detailed sales for that product.                    |

13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction

| Point               | Explanation                                                                                                                                                                                                           |
| ------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| 1. Definition       | **Keep All Filters** is an option in Power BI Drill-Through that keeps all the filters applied on the source page when navigating to the drill-through page.                                                          |
| 2. Purpose          | It ensures that the detailed report page shows data based on the same filter conditions selected by the user.                                                                                                         |
| 3. User Interaction | When a user drills through a data point, all slicers and filters (like region, category, or date) remain applied on the new page.                                                                                     |
| 4. Effect on Data   | The drill-through page displays more accurate and context-specific data because it respects the previous filters.                                                                                                     |
| 5. Example          | If a user filters **Region = North** and **Category = Electronics** on the main page and drills through a product, the detailed page will show only **Electronics sales in the North region** instead of all regions. |
