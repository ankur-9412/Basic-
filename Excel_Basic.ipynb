    A) Relative Cell Reference (e.g., A2, B2)
       Meaning:
           Changes automatically when the formula is copied to another cell.
           Default referencing type in Excel.
          ** example:**
           =B2 * 10%
           and copy it down, Excel changes it to:

          In C3 → =B3 * 10%

          In C4 → =B4 * 10%
      B) Absolute Cell Reference (e.g., $B$1)
         Meaning:
              Does not change when copied.
              A $ sign locks both row and column.
             Example:
             =B2 * $E$1
             =B3 * $E$1
             =B4 * $E$1

      C) Mixed Cell Reference (e.g., $B2 or B$2)
         Meaning:
               Either row or column is locked.
               Useful when copying formulas across rows and columns.
               Example:
               =$B2 * C1
