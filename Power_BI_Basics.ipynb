Question 1: What is Power BI and why is it used in businesses?


```
Power BI is a Business Intelligence (BI) tool developed by Microsoft that helps organizations analyze data and create interactive reports and dashboards.

Why it is used in businesses:
 > To convert raw data into meaningful insights
 > To create interactive dashboards and visual reports
 > To support data-driven decision making
 > To track KPIs (Key Performance Indicators)
 > To combine data from multiple sources
```
Question 2 : Name and explain the three main components of Power BI.


```
1. Power BI Desktop
    > A Windows application used to create reports and dashboards.
    > Used for data modeling, data cleaning, and visualization creation.

2. Power BI Service (PowerBI.com)
    > A cloud-based service.
    > Used to publish, share, and collaborate on reports online.

3. Power BI Mobil
    > Mobile application for Android and iOS.
    > Allows users to view reports and dashboards on smartphones.
```
Question 3 : Explain the Power BI workflow.


```
The Power BI workflow follows these steps:

  1. Connect Data – Import data from Excel, CSV, Database, etc.
  2. Transform Data (Power Query) – Clean and prepare data.
  3. Model Data – Create relationships between tables.
  4. Create Visualizations – Design charts, tables, dashboards.
  5. Publish Report – Upload report to Power BI Service.
  6. Share & Collaborate – Share dashboards with team members.
```

Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.


```
1. Data View
   > Shows data in table format (rows and columns).
   > Used to inspect and verify data.

2. Report View
   > Default view for creating visualizations.
   > Used to build charts, tables, cards, maps, dashboards.

3. Model View
   > Shows relationships between tables.
   > Used to create and manage relationships.
```

Question 8 : Split Owner Name to create two new columns as First Name and Last Name


```
Steps in Power Query:
  1. Click Transform Data
  2. Select the Owner Name column
  3. Click Split Column
  4. Choose By Delimiter5
  5. Select delimiter as Space
  6. Click OK

Two new columns will be created:
  > Owner Name.1 → Rename to First Name
  > Owner Name.2 → Rename to Last Name

If names contain middle names, use:
  > Split by space
  > Choose “Split into Columns”
  > Adjust accordingly
```
