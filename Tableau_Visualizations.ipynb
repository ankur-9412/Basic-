Q9. Chart Selection Justification

Explain why different charts were used for different business questions in this analysis.



```
Different charts were used in this analysis because each business question requires a specific type of visualization to present data clearly and effectively.

 Explanation:
  > Bar Chart (Category & Region Analysis):
     > Used to compare values across categories or regions. It makes it easy to identify highest and lowest performers.

  > Line Chart (Sales Trend & Growth):
    > Used for time-based data to show trends and patterns over time, helping in identifying growth, peak, and slow periods.

  > Pie Chart (Contribution Analysis):
    > Used to show proportion or percentage contribution of each region to total sales.

  > Scatter Plot (Sales vs Profit):
    > Used to analyze the relationship between two variables and identify loss-making or low-profit orders.

  > Bubble Chart (Product Prioritization):
    > Used for multi-dimensional analysis by combining sales, profit, and quantity in a single view.

  > Dashboard:
    > Combines multiple charts to provide a complete and interactive overview for decision-making.
```
