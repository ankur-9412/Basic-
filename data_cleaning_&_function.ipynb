1. What is data cleaning, and why is it important in data analysis?
*   What are the potential consequences of analyzing unclean or messy data?
*   Explain the common steps involved in cleaning and organizing data.
  
        -> Data cleaning means identifying and fixing or removing incorrect, corrupted, duplicate, or incomplete data (e.g., filling or removing missing values, removing duplicates, standardizing formats). It's essential because dirty data leads to biased or wrong analysis, poor model performance, and unreliable business decisions.

        -> 1. Misleading conclusions and bad insights.
        2. Incorrect reports and charts.
        3. Poor ML/statistical model performance.
        4. Bad business decisions and possible financial losses.
        5. Wasted time and resources cleaning up later.

        -> 1. Inspect the data (columns, types, sample rows)
           2. Make a backup copy.
           3. Identify missing values.
           4. Handle missing values (remove, fill with mean/median/mode, or impute).
           5. Remove duplicates.
           6. Fix types/formats (dates, numbers), and trim text.
           7. Detect and verify or handle outliers.
           8. Standardize units, names, categories.
           9. Apply validation rules (e.g., Data Validation).
           10. Document changes and rules applied.

3.Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

     -> * TRIM(text) removes leading/trailing and extra internal     spaces — good for cleaning names/addresses.
        * LEFT(text, n) extracts leftmost n characters — useful for prefixes/codes.
        * RIGHT(text, n) extracts rightmost n characters — useful for file extensions or suffixes.
        * MID(text, start, n) extracts a middle substring starting at position start for n characters.
        * CONCAT(...) or text1 & text2 joins text pieces — e.g., combine first and last name.

8. Understanding TODAY() and NOW()

 a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use
 each function.
   
       -> TODAY() returns today's date only (time = 00:00). NOW() returns current date and time. Use TODAY() for daily reports, NOW() for timestamping events.
