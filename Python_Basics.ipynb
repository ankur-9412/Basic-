Question 2 : Explain the difference between List and Tuple in Python.


```
| Feature         | List                       | Tuple                            |
| --------------- | -------------------------- | -------------------------------- |
| **Mutability**  | Mutable (can be changed)   | Immutable (cannot be changed)    |
| **Syntax**      | Uses square brackets `[ ]` | Uses parentheses `( )`           |
| **Performance** | Slower compared to tuple   | Faster than list                 |
| **Usage**       | Used when data may change  | Used when data should not change |
| **Example**     | `[1, 2, 3]`                | `(1, 2, 3)`                      |

```
Question 3 : What is a function in Python? Why are functions useful?


```
A function in Python is a block of reusable code that performs a specific task. It is defined using the def keyword and can be called whenever needed.

Why are functions useful?

1. Code Reusability
   > Functions allow you to reuse the same code multiple times without rewriting it.

2. Better Organization
   > They help divide a program into smaller, manageable parts, making it easier to understand.

3. Easy Maintenance
   > Changes can be made in one place (inside the function) instead of modifying the entire code.

4. Reduces Errors
   > Since repetition is reduced, the chances of errors are also minimized.
```

````python
# Original list with duplicates
numbers = [1, 2, 2, 3, 4, 4, 5]

# Removing duplicates using set
unique_numbers = list(set(numbers))

# Printing the result
print("List after removing duplicates:", unique_numbers)
````

````python
# Given list
numbers = [2, 3, 4, 2, 5, 2]

# Taking input from the user
num = int(input("Enter the number to count: "))

# Counting occurrences
count = numbers.count(num)

# Printing the result
print("The number appears", count, "times in the list")
````
