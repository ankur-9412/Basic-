Q1. What is the average, median, and standard deviation of house prices?

```
import pandas as pd

# Load dataset
df = pd.read_csv('Downloads/House Price India.csv')

# Basic checks
print(df.info())
print(df.describe())

# Check missing values
print(df.isnull().sum())

# Handle missing values
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Q1. Average, Median, and Standard Deviation of House Prices

average_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()

print("Average House Price:", average_price)
print("Median House Price:", median_price)
print("Standard Deviation of House Price:", std_price)



Explanation
Mean (Average)
  -> It shows the overall average price of houses in the dataset.

Median
  -> It represents the middle house price when all prices are arranged in order.
  -> Median is useful because it is less affected by very high or very low prices.

Standard Deviation
  -> It measures how much the house prices vary from the average price.

   High standard deviation → prices vary a lot

   Low standard deviation → prices are more consistent

Interpretation
  -> If the average price is greater than the median price, it means there are some expensive houses increasing the average.

  -> A large standard deviation indicates a wide price range in the real estate market.

```

Q3. Check if the price data is skewed.



```
# Calculate skewness
price_skewness = df['Price'].skew()

print("Skewness of House Prices:", price_skewness)

Skewness Value Meaning
  -> Skewness = 0 → Perfectly normal distribution

  -> Skewness > 0 → Positively skewed (right-skewed)

  -> Skewness < 0 → Negatively skewed (left-skewed)
```

Q6. Identify any anomalies where houses have high prices but low areas.



```
high_price = df['Price'].quantile(0.75)   # Top 25% prices
low_area = df['living area'].quantile(0.25)  # Bottom 25% areas

# Filter anomalies
anomalies = df[(df['Price'] > high_price) &
               (df['living area'] < low_area)]

# Display anomalies
print(anomalies)
```

Q9. Which location (zipcode/area) has the highest average price?



```
# Average price by zipcode
avg_price_location = df.groupby('Postal Code')['Price'].mean()

# Find location with highest average price
highest_price_location = avg_price_location.idxmax()
highest_avg_price = avg_price_location.max()

print("Location with Highest Average Price:", highest_price_location)
print("Highest Average Price:", highest_avg_price)
```
