#Theoretical Questions :
Question 1 :

What is a Calculated Field and why is it important in business dashboards?


```
A Calculated Field in Tableau is a custom field created using formulas to perform calculations on existing data.

Importance in business dashboards:
  > Helps in creating new metrics (e.g., Profit Ratio = Profit / Sales)
  > Enables better insights without modifying original data
  > Supports dynamic analysis (like conditional logic, KPIs)
  > Makes dashboards more meaningful and decision-driven

 Example:
Instead of just showing sales, you can create a calculated field to show Profit Margin, which is more useful for management.
```
Question 2 :

 Why can’t parameters change visuals directly in Tableau?



```
Parameters in Tableau are independent variables that do not directly interact with visualizations.

Reason:

  > Parameters don’t automatically connect to charts or data
  > They need to be used in:
    > Calculated fields
    > Filters
    > Reference lines

 So, parameters act like input controls, but to impact visuals, they must be linked through logic (calculated fields).
```
Question 3 :

When should EXCLUDE LOD be used?


```
EXCLUDE LOD (Level of Detail) is used when you want to remove a specific dimension from the view’s level of detail.

Use case:
  > When you want to aggregate data at a higher level than what is shown in the view

Example:
If your view shows Sales by Region and Category, but you want total sales only by Region (ignoring Category), you use EXCLUDE.
```
Question 4 :

Difference between FIXED and INCLUDE LOD expressions?


```
| Feature    | FIXED LOD                             | INCLUDE LOD                          |
| ---------- | ------------------------------------- | ------------------------------------ |
| Definition | Calculates at a fixed level of detail | Adds extra dimension to current view |
| Dependency | Independent of view                   | Dependent on view                    |
| Control    | More control over aggregation         | Flexible with view changes           |
| Use case   | When you want constant values         | When you want to refine calculation  |

```
