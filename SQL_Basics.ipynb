#Q1. Create a New Database and  Table for Employees

Task: Create a new database named company_db and Create a table named employees with the following
columns:

```

CREATE DATABASE company_db;


USE company_db;

 Create employees table
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);

```

#Q9. What is the Difference Between a Primary Key and a Foreign Key?


```
| Aspect              | Primary Key                                                                  | Foreign Key                                                              |
| ------------------- | ---------------------------------------------------------------------------- | ------------------------------------------------------------------------ |
| **Definition**      | A column (or set of columns) that uniquely identifies each record in a table | A column that creates a link between two tables                          |
| **Uniqueness**      | Must contain **unique** values                                               | Can contain **duplicate** values                                         |
| **NULL values**     | **NOT allowed**                                                              | **Allowed** (unless restricted)                                          |
| **Purpose**         | Ensures each record is uniquely identifiable                                 | Maintains **referential integrity**                                      |
| **Location**        | Defined in the **same table**                                                | Defined in the **child table**, refers to a primary key in another table |
| **Count per table** | Only **one** primary key per table                                           | A table can have **multiple** foreign keys                               |
| **Example**         | `employee_id` in `employees` table                                           | `employee_id` in `sales` table referencing `employees`                   |

```
#Q10. What Are Constraints in SQL and Why Are They Used?


```
| Constraint      | Description                               | Example                                             |
| --------------- | ----------------------------------------- | --------------------------------------------------- |
| **PRIMARY KEY** | Uniquely identifies each record; no NULLs | `employee_id PRIMARY KEY`                           |
| **FOREIGN KEY** | Links one table to another                | `FOREIGN KEY (emp_id) REFERENCES employees(emp_id)` |
| **NOT NULL**    | Prevents NULL values                      | `name VARCHAR(50) NOT NULL`                         |
| **UNIQUE**      | Ensures all values are unique             | `email UNIQUE`                                      |
| **CHECK**       | Limits value range                        | `salary CHECK (salary > 0)`                         |
| **DEFAULT**     | Assigns a default value                   | `status DEFAULT 'Active'`                           |

```
