1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?


```
Tableau Public allows dashboards to be shared publicly with no data security, whereas Tableau Server/Cloud restricts access to authorized users and ensures data security through authentication and permissions.

 > Tableau Public → Open for all (no security)
 > Tableau Server/Cloud → Restricted access (secure)
```
2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?


```
Live Connection:
 > Connects directly to the original data source
 > Fetches real-time data whenever the dashboard is opened or used.
 > No need for manual refresh.
 > Performance depends on database speed and network; may be slower.

Extract:
 > Stores a snapshot (copy) of the data inside Tableau.
 > Requires manual or scheduled refresh to update data.
 > Faster performance due to optimized storage
 > Does not show real-time data unless refreshed.

```
3. In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?


```
In Tableau Server / Tableau Cloud:
 > Permissions and User Roles control access
 > Admins can define:
     > Who can view dashboards
     > Who can interact (filters, drill-downs)
     > Who can edit or publish content

 >Permissions can be set at:
    > Project level
    > Workbook level
    > Data source level
```
4. If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server?



```
To prevent users from downloading data:
  > Adjust Permissions settings.
  > Disable:
    > Download Full Data
    > Download Summary Data

  > Can also restrict:
    > Export to Excel / PDF

 This ensures users can view dashboards but cannot access raw data, improving security and privacy.
```
5. Name two methods by which a Tableau dashboard can be embedded into a web page or website.


```
Two common methods:
   > Embed Code (iframe):
      > Copy embed link and paste into website HTML.
      > Simple and widely used.

   > Tableau JavaScript API:
      > Allows advanced interaction and customization.
      > Developers can control filters, events, and behavior.

 Both methods help integrate dashboards into web pages.
```
6. List two best practices you should follow when preparing a dashboard for public sharing to ensure data
security and privacy.



```
To ensure security and privacy:
   > Remove or avoid sensitive/confidential data.
   > Use data anonymization or aggregation.> > Hide unnecessary fields and filters.
   > Use extracts instead of live connections (if needed).
   > Test dashboard before publishing.

```
7. After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause



```
Most common reason:
  > Broken data source connection

Possible causes:
   > File path changed or file deleted.
   >  Database credentials missing/expired.
   > Data source not published with workbook.> Server cannot access local files.

 Fix by updating connection, republishing data source, or correcting credentials.
```
